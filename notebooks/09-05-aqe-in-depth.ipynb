## AQE in depth — runtime re-optimization

Module 02 introduced AQE with **dynamic coalescing** of post-shuffle partitions. Two more behaviors complete the picture — and both fire *after* a shuffle's real statistics are known:

- **Dynamic join strategy switch** — Spark can promote a sort-merge join to a **broadcast hash join** at runtime if one side turns out small enough (threshold: `spark.sql.adaptive.autoBroadcastJoinThreshold`).
- **Skew join handling** — Spark detects an oversized partition and **splits it** into sub-tasks so one slow task can't stall the stage. A partition counts as skewed when it's both ≥ 5× the median (`skewedPartitionFactor`) **and** ≥ 256 MB (`skewedPartitionThresholdInBytes`).

The theme is the same as module 02: the **same Catalyst optimizer**, given a second chance at each stage boundary with *real* numbers. On by default since Spark 3.2 — it's why so much manual tuning is now automatic.
