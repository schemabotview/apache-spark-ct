## Why Spark Connect exists

The classic single-process model has friction in four places, and Connect was designed to fix each one:

- **Shared JVM.** Client and driver share one JVM, so you hit version mismatches, dependency conflicts, and classpath surprises. With Connect the client is a thin library and the server owns all JVM dependencies.
- **JVM-only languages.** Every language needed a Py4J-style bridge to talk to Spark. Connect speaks gRPC, so *any* language with a gRPC stub — Go, Rust, Node — can be a client.
- **Fragile dev sessions.** Restarting a notebook killed the driver. Now the server lives independently; clients connect and disconnect freely.
- **Hard upgrades.** Every client had to match the cluster's Spark version. Client and server can now differ, within protocol compatibility.

You start a server with `start-connect-server.sh` and point a client at it with `SparkSession.builder.remote("sc://host:15002")` — or the `SPARK_REMOTE` environment variable, which `getOrCreate()` picks up automatically. Spark Connect became Generally Available in **Spark 3.5**; **Databricks Connect** is built on top of it.
