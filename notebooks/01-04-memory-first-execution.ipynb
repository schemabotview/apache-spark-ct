## Memory-first execution

The single design choice that makes Spark fast is this: keep intermediate results in executor RAM instead of writing them to disk between every step.

The previous generation, Hadoop MapReduce, did a full disk round-trip after every stage — write the intermediate output to disk, then read it back for the next step. On a multi-step pipeline, that disk traffic dominated the runtime. Spark keeps the data in memory and hands it straight to the next step; data only spills to disk when something forces it out — memory pressure, or an explicit checkpoint.

That one decision is most of Spark's speed advantage over MapReduce. Everything else you'll learn later — caching, broadcast joins, adaptive query execution — is an optimization layered on top of this memory-first core.

The trade-off is that memory is finite: a job that tries to hold too much spills to local disk, and in the worst case fails with an out-of-memory error. Sizing partitions so they fit comfortably in executor memory is a recurring theme once we reach performance tuning.
