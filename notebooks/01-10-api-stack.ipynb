## RDD vs DataFrame vs Dataset — the API stack

Spark exposes three layered APIs over the same engine. Knowing which to reach for is a common exam question.

- **RDD** — the original, low-level API: a distributed collection of objects with full control and no built-in schema. Powerful, but you get **no Catalyst optimization** — Spark can't see inside your lambdas.
- **DataFrame** — a distributed table with named, typed columns. The API you'll use almost always. Because operations are column expressions, **Catalyst can optimize them** — and in Python, DataFrames are the only structured API (there is no typed Dataset in PySpark).
- **Dataset** — a **JVM-only** (Scala/Java) typed view over a DataFrame, adding compile-time type safety. It does not exist in Python.

The practical rule: **default to DataFrames.** They're optimized, concise, and portable across languages. Drop to RDDs only when you need low-level control the DataFrame API can't express — custom partitioning, or non-tabular data. In PySpark, "DataFrame vs Dataset" isn't a real choice; the real choice is DataFrame vs RDD, and it's almost always DataFrame.
