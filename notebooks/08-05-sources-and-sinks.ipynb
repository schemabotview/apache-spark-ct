## Sources & sinks

A **source** is a pipeline's entry point; Spark tracks its progress (offsets / file lists) in the checkpoint so each micro-batch is processed exactly once on restart.

- **Rate** (`"rate"`) — testing / benchmarking. **File** (`json` / `csv` / `parquet`) — file-drop ETL; needs an **explicit schema** and options like `maxFilesPerTrigger`, `cleanSource`. **Kafka** (`"kafka"`) — production event streaming (`subscribe`, `startingOffsets`, `maxOffsetsPerTrigger`). **Socket** — demos only, **not** fault-tolerant.

A **sink** is where each micro-batch's output goes — sinks differ in supported output modes and durability:

- **memory** / **console** — dev / debug, not fault-tolerant. **file** — append-only, fault-tolerant. **Delta** — the **production default**: all output modes, exactly-once via transactional commits, no small-file problem. **Kafka** — event forwarding. **`foreachBatch`** — the escape hatch: receives the micro-batch as a regular DataFrame + a `batch_id` for multi-table writes or a non-Spark API call.
