## Persistence — `cache()`, `persist()`, StorageLevel

By default, **every action re-executes the entire lineage from the source.** Call `.count()` and then `.collect()` on the same RDD and you just computed it twice.

`.cache()` materializes the RDD on the first action and reuses it after. `.persist(level)` lets you choose the storage policy:

- **`MEMORY_ONLY`** — executor RAM; the default for `.cache()` on RDDs. Recompute on a miss.
- **`MEMORY_AND_DISK`** — RAM if it fits, spill to disk otherwise; the default for `.cache()` on DataFrames.
- **`DISK_ONLY`** — always to disk. Cheap on memory, slow to read.
- **`*_SER`** — store serialized bytes: less RAM, more CPU.
- **`*_2`** — replicate each partition on two executors for resilience.

Call **`.unpersist()`** when you're done — a cache holds executor memory until you release it.
